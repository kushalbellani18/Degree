## Tokenization

```python
sent_tokens = nltk.sent_tokenize(text)
word_tokens = nltk.word_tokenize(text)
```

## Preprocessing

```python
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()

removePunctDict = dict(
    (ord(p), None) for p in string.punctuation
)

def lemTokens(tokens):
    return [lemma.lemmatize(token) for token in tokens]

def lemNormalize(text):
    w = nltk.word_tokenize(
## translate() function is very bright to change from one character to another with the help of ascii, which store in dict variable.
        text.lower().translate(removePunctDict)  
    )

    return lemTokens(w)
```

## Greetings

```python
greetingInputs = ("hello", "hi", "greetings", "sup", "what's up", "hey")
greetingOutputs = [
    "hi", "hey", "nods", "hi there", "hello", "I am glad! you can ask me"
]

def greeting(sentence):
    for word in sentence.split():
        if word.lower() in greetingInputs:
            return random.choice(greetingOutputs)
```

## Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def response(user):
    bot = ''
    
    sent_tokens.append(user)
    TfidfVec = TfidfVectorizer(tokenizer=lemNormalize, stop_words="english")
    tfidf = TfidfVec.fit_transform(sent_tokens)
    vals = cosine_similarity(tfidf[-1], tfidf)
    idx = vals.argsort()[0][-2]
    flat = vals.flatten()
    flat.sort()
    req_tfidf = flat[-2]
    
    if(req_tfidf == 0):
        bot += "I am sorry! What do you say meaning?"
        return bot
    else:
        bot += sent_tokens[idx]
        return bot
```
