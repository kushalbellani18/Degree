```python
import pandas as pd
```

```python
data = [
    ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
    ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
    ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice Cream', 'Eggs']
]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
test = TransactionEncoder()
```

```python
test1 = test.fit(data).transform(data)
test1
```

```python
test1.astype(int)
```

```python
test.columns_
```

```python
df = pd.DataFrame(test1, columns=test.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
dfA = apriori(df, min_support=0.5, use_colnames=True)
dfA
```

```python
dfA['length'] = dfA['itemsets'].apply(lambda x: len(x))
dfA
```

```python
dfA['lengthForTwo'] = dfA['itemsets'].apply(lambda x: len(x) == 2)
dfA
```

```python
lol = dfA[['itemsets', 'length']].where(dfA.length == 2).dropna()
lol
```

```python
dfA[dfA['length'] == 3]
```

```python
dfA[(dfA['length'] == 2) & (dfA['support'] >= 0.6)]
```
