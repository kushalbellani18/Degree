## 6) Plot a word cloud that shows all the words that are most frequently used in positive and negative tweets.

```python
import wordcloud

def showWC(df):
    txt = ' '.join(df['tweet_text'].astype(str).tolist())
    fig = wordcloud.WordCloud(background_color="black", colormap="Reds", width=1900, height=1300).generate(txt)
    
    plt.imshow(fig)
```

```python
showWC(df[df['sentiment'] == "positive"])
```

```python
showWC(df[df['sentiment'] == "negative"])
```

```python
import re

def selectRT(lol):
    txt = ' '.join(lol.astype(str).tolist())
    
    x = re.findall("RT\s+", txt)
    print(x)
    
def selectAT(lol):
    txt = ' '.join(lol.astype(str).tolist())
    
    x = re.findall("\B@\w+", txt)
    print(x)
    
def selectHash(lol):
    txt = ' '.join(lol.astype(str).tolist())
    
    x = re.findall("\B#\w+", txt)
    print(x)
    
def removeAT(lol):
    #txt = ' '.join(lol.astype(str).tolist())
    
    return re.sub("\B@\w+", "", lol)

def removeURL(lol):
    return re.sub(r'http\S+', "", lol)

def removeHash(lol):
    return re.sub("\B#\w+", "", lol)
```

## 11) Perform tokeniztion on some raw text and display statements as well as words
