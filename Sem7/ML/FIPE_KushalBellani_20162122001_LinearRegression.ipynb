```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset/vechiclePrice.csv")

df.head()
```

```python
df.info()
```

```python
df['name'].unique()
```

Problem is name; Here, better get company name instead of model name.

```python
for n in range(0, len(df['name'])):
    companyName = df['name'][n].split(' ')[0]
    
    df['name'][n] = companyName
```

```python
df.head()
```

## Do Data Transformation

```python
for n in range(0, len(df['name'].unique())):
    df['name'].replace(df['name'].unique()[n], n, inplace=True)
    
for n in range(0, len(df['year'].unique())):
    df['year'].replace(df['year'].unique()[n], n, inplace=True)
    
for n in range(0, len(df['fuel'].unique())):
    df['fuel'].replace(df['fuel'].unique()[n], n, inplace=True)
    
for n in range(0, len(df['seller_type'].unique())):
    df['seller_type'].replace(df['seller_type'].unique()[n], n, inplace=True)
    
for n in range(0, len(df['transmission'].unique())):
    df['transmission'].replace(df['transmission'].unique()[n], n, inplace=True)
    
for n in range(0, len(df['owner'].unique())):
    df['owner'].replace(df['owner'].unique()[n], n, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

## Apply Modeling (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df[['name', 'year', 'km_driven', 'fuel', 'seller_type', 'transmission', 'owner']]
y = df['selling_price']
```

```python
lr = LinearRegression()

lr.fit(x, y)
```

```python
yHat = lr.predict(x)
```

```python
for i in range(0, 5):
    print(yHat[i], " vs ", y[i])
```

## Apply MSE and R2_Score

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("MSE: ", mean_squared_error(yHat, y))
print("r2_score: ", r2_score(yHat, y))
```

```python
df = pd.read_csv("dataset/vechiclePrice.csv")

df.head()
```

```python
for n in range(0, len(df['name'])):
    companyName = df['name'][n].split(' ')[0]
    
    df['name'][n] = companyName
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
df.info()
```

```python
x = df[[
    *(c for c in df.columns if c != 'selling_price')
]]
y = df['selling_price']

lr = LinearRegression()
lr.fit(x, y)

yHat = lr.predict(x)

for i in range(0, 5):
    print(yHat[i], " vs ", y[i])
```

```python
print("MSE: ", mean_squared_error(yHat, y))
print("r2_score: ", r2_score(yHat, y))
```
