# Feature extraction using Correlation

## For Numerical

```python
from scipy.stats import pearsonr
```

```python
numLst = ["temp", "rain_1h", "snow_1h", "clouds_all"]
tar = "traffic_volume"

resLst = []
```

```python
for nl in numLst:
    corr, _ = pearsonr(df[tar], df[nl])
    
    if corr > 0.5 or corr < -0.5:
        resLst.append(
            [nl, corr, "Success"]
        )
        
    else:
        resLst.append(
            [nl, corr, "Fail"]
        )
```

```python
resLst = pd.DataFrame(resLst)
resLst.columns = ["Attribute", "Correlation", "Status"]

resLst
```

## For Categorical

```python
from scipy import stats
```

```python
catLst = ['holiday', 'weather_main', 'weather_description', 'date_time']

resLst = []
```

```python
for lol in catLst:
    res = [lol]
    
    groupedTest = df[[lol, tar]].groupby([lol])
    unique_ = df[lol].unique()
    
    f_val, p_val = stats.f_oneway(
        *(groupedTest.get_group(value)[tar] for value in unique_)
    )
    
    res.append(f_val)
    res.append(p_val)
    
    dfn = len(groupedTest) - 1
    dfd = 0
    for value in unique_:
        dfd += groupedTest.get_group(value)[tar].shape[0]
        
    res.append(
        stats.f.ppf(q=1-0.05, dfn=dfn, dfd=dfd)
    )
    
    resLst.append(res)
```

```python
resLst = pd.DataFrame(resLst)
resLst.columns = ["Attribute", "F-test", "P-value", "Results"]

resLst
```

# Modeling

## Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
x = df[['temp']]
y = df[tar]

lm.fit(x, y)
```

```python
print("Intercept: ", lm.intercept_, " whereas, slope: ", lm.coef_)
```

Thus, final extimated model in traffic_volume = -2198.1868 + (19.4093) * temp

```python
yHat = lm.predict(x) 

print("Predicated Value\tActual Value\n")
for i in range(0, 5):
    print(yHat[i], "\t", y[i])
```

## Multiple Linear Regression

```python
zList = []

for b in list(df.columns):
    if b != tar and b != "date_time":
        zList.append(b)
        
z = df[[
    *(x for x in zList)
]]
```

```python
lm = LinearRegression()

lm.fit(z, df[tar])
```

```python
print("Intercept: ", lm.intercept_, " whereas, slope: ", lm.coef_)
```

```python
yHatMulti = lm.predict(z)

print("Predicated value\tActual value\n")

for i in range(0, 5):
    print(yHatMulti[i], "\t", y[i])
```

## Polynomial Regression

```python
x = df['temp']
y = df[tar]

f = np.polyfit(x, y, 3)
p = np.poly1d(f)

print(p)
```

```python
yPolyPred = p(x)

print("Predicated\tActual\n")
for i in range(0, 5):
    print(yPolyPred[i], "\t", y[i])
```

## Multiple Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
ipt = [('Scale', StandardScaler()), ('Polynomial', PolynomialFeatures(2)), ('Model', LinearRegression())]

pipe = Pipeline(ipt)
pipe
```

```python
pipe.fit(z, y)
```

```python
yPipe = pipe.predict(z)

print("Predicated\tActual\n")
for i in range(0, 5):
    print(yPipe[i], "\t", y[i])
```

# MSE and R^2

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print("Simple Linear Regression")
print("R^2: ", r2_score(df[tar], yHat))
print("MSE: ", mean_squared_error(df[tar], yHat))

print("-----------------------------------------------------------------")

print("Multiple Linear Regression")
print("R^2: ", r2_score(df[tar], yHatMulti))
print("MSE: ", mean_squared_error(df[tar], yHatMulti))

print("-----------------------------------------------------------------")

print("Simple Polynomail Regression")
print("R^2: ", r2_score(df[tar], yPolyPred))
print("MSE: ", mean_squared_error(df[tar], yPolyPred))

print("-----------------------------------------------------------------")

print("Mulitple Polynomial Regression")
print("R^2: ", r2_score(df[tar], yPipe))
print("MSE: ", mean_squared_error(df[tar], yPipe))
```
